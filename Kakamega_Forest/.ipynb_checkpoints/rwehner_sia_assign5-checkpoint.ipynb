### Part 1--Get image from Sentinelsat that includes Harrisburg, PA

```python
# get packages necessary to utilize Sentinelsat API
import sentinelsat
from sentinelsat import SentinelAPI, read_geojson, geojson_to_wkt
from datetime import date
# creditials for API
api = SentinelAPI(
    'rwehner', 
    '7yN@*bWEd518', 
    'https://apihub.copernicus.eu/apihub'
)

# create geojson encompassing the Harrisburg, PA from geojson.io 
hbg_geometry = {
  "type": "Polygon",
  "coordinates": [
    [ 
      [-76.937255859375, 40.23183929314176],
      [-76.8445587158203, 40.23183929314176],
      [-76.8445587158203, 40.29262043670974],
      [-76.937255859375, 40.29262043670974],
      [-76.937255859375, 40.23183929314176]
    ]
  ]
}

# convert geojson into well-known-text so that the api can be queried
hbg_footprint = geojson_to_wkt(hbg_geometry)
hbg_footprint
```

```python
# build the query for the api
hbg_image_metadata = api.query(
    hbg_footprint,
    date=('20220901', '20220930'),    #date range 9/01/22-9/30/22
    platformname='Sentinel-2',
    processinglevel='Level-2A'         #separate 1C(topofatmosphere) from 2A(surfacereflectance)
)
hbg_image_metadata
```

```python
# examine metadata of queried images utilizing panda to create dataframe to csv
import pandas as pd
hbg_image_metadata_df = api.to_dataframe(hbg_image_metadata)
hbg_image_metadata_df.to_csv('hbg_metadata.csv')

# sort images by cloud cover 
hbg_image_metadata_df_sorted = hbg_image_metadata_df.sort_values(
    ['cloudcoverpercentage'], ascending=[True])

# knowing my goal is to find a corresponding Planet image with the closest date match, 
# create a subset of images with the least cloud in order to have a higher match potential with Planet 
hbg_image_metadata_df_sorted.to_csv('hbg_metadata_sorted.csv')
hbg_image_metadata_df_sorted = hbg_image_metadata_df_sorted.head(2)
len(hbg_image_metadata_df_sorted)    #double checking how many images are in subset
```

```python
# download images with the least cloud cover
api.download_all(hbg_image_metadata_df_sorted.index)
```

```python
# file management of downloaded zips
import os   
import zipfile

# find the downloaded file names, create a list, and add the only the zip files using a loop
all_filenames_in_folder = os.listdir()
hbg_filenames_to_unzip = []
for filename in all_filenames_in_folder:
    if filename.endswith('.zip'): 
        hbg_filenames_to_unzip.append(filename)
        
# Create a folder for the images and add the unzipped files, remove the zip for a clean workspace
hbg_folder = 'hbg_unzipped'
if not os.path.exists(hbg_folder):
    os.mkdir(hbg_folder)    #create the folder
for filename in hbg_filenames_to_unzip:
    with zipfile.ZipFile(filename, 'r') as zip_ref:
        zip_ref.extractall(hbg_folder)
    os.remove(filename)
```

### Part 2--Get image from Planet API that includes Harrisburg, PA

```python
import planet
# assign Planet API key as an environment variable and test by printing
os.environ['PL_API_KEY'] = 'PLAKaf8ea4b97c8a44119b26bc4d962de4bb' 
print(os.getenv("PL_API_KEY"))
```

```python
# create the filters needed to find images on the Planet API

# find images that intersect with Harrisburg geojson 
geometry_filter = {
  "type": "GeometryFilter",
  "field_name": "geometry",
  "config": hbg_geometry 
}

# find images that correspond with the Sentinel-2 images 
date_range_filter = {
  "type": "DateRangeFilter",
  "field_name": "acquired",
  "config": {
    "gte": "2022-09-19T00:00:00.000Z",
    "lte": "2022-09-21T00:00:00.000Z"
  }
}

# filter to only return cloud cover less than 20%
cloud_cover_filter = {
  "type": "RangeFilter",
  "field_name": "cloud_cover",
  "config": {
    "lte": 0.20
  }
}

# concatenate into one filter
combined_filter = {
  "type": "AndFilter",
  "config": [geometry_filter, date_range_filter, cloud_cover_filter]
}

combined_filter    #confirm that it is working
```

```python
# get set for "talking" to the Planet API by getting necessary packages
import json
import requests
from requests.auth import HTTPBasicAuth

# assign API Key
PLANET_API_KEY = os.getenv('PL_API_KEY')
item_type = "PSScene"

# define the object to request from the API
hbg_search_request = {
  "item_types": [item_type], 
  "filter": combined_filter
}
hbg_search_request    #double checking that all the info is there
```

```python
# send POST request to the Planet API for images using url, api key, and combined filter
hbg_result = requests.post(
    'https://api.planet.com/data/v1/quick-search',
    auth=HTTPBasicAuth(PLANET_API_KEY, ''),
    json=hbg_search_request)
hbg_result 
```

```python
results = hbg_result.json()
print(results)
```

```python
# create new shorter list showing just image id's as it is hard to read the full json above
image_ids = [feature['id'] for feature in results['features']]
print(image_ids)
id0 = image_ids[0]
id0
```

```python
# creating the path needed for downloading by inserting the image into the url 
id0_url = 'https://api.planet.com/data/v1/item-types/{}/items/{}/assets'.format(item_type, id0)

# rewriting the request to the API for the needed image
result = requests.get(
    id0_url,
    auth=HTTPBasicAuth(PLANET_API_KEY, '')
  )
result
```

```python
#looking at the type of images available for download
print(result.json().keys())
# choosing the ortho 8b and checking to see if the file is active or inactive
print(result.json()['ortho_analytic_8b']['status'])
# getting the needed pieces to put the activation request together
links = result.json()[u"ortho_analytic_8b"]["_links"]
self_link = links["_self"]
activation_link = links["activate"]
activation_link
```

```python
# send the activation request to the api
activate_result = requests.get(
    activation_link,
    auth=HTTPBasicAuth(PLANET_API_KEY, '')
  )
activate_result
```

```python
# check the progress of the activation...waiting for activating to turn to activate
activation_status_result = requests.get(
    self_link,
    auth=HTTPBasicAuth(PLANET_API_KEY, '')
  )
    
print(activation_status_result.json()["status"])
```

```python
# get link to download tiff
download_link = activation_status_result.json()["location"]
print(download_link)
```

### Part 3--Clipping AOI

```python
# to work with raster images, get necessary packages
import rasterio 
from matplotlib import pyplot as plt 
import numpy as np 

#specify filenames and assign as a rasterio object
hbg_planet_file = "hbg20220919_153015_80_2473_3B_AnalyticMS_8b.tif"
hbg_s2_file = "T18TUK_20220920T155941_TCI_10m.jp2"
ro_planet = rasterio.open(hbg_planet_file)
ro_s2 = rasterio.open(hbg_s2_file)

#scale values for display purposes with pyplot
def scale(band): 
    return band / 10000.0

# make a true color image from the 8 band planet image
# Sentinel image is already a true color image
blue_p = scale(ro_planet.read(2))
green_p = scale(ro_planet.read(4))
red_p = scale(ro_planet.read(6))
ro_planet_tci = np.dstack((red_p, green_p, blue_p))
#display
plt.imshow(ro_planet_tci)
```

```python
# define AOI with geojson of Harrisburg, PA
# choose this method over dividing the x and y range to ensure Harrisburg AOI is included 
# ensuring continuity between varying sizes of Planet and Sentinel images

hbg_aoi_geometry = [{
  "type": "Polygon",
  "coordinates": [
    [ 
      [336047.96, 4456362.46],
      [343349.23, 4456207.26],
      [343463.15, 4461688.80],
      [336167.19, 4461844.04],
      [336047.96, 4456362.46]
    ]
  ]
}]

hbg_aoi_geometry   
```

```python
# Understanding my Planet image to help problem solve errors

print(ro_planet.name)
print(ro_planet.tags())
print(ro_planet.count)
ro_planet.crs
ro_planet.profile
```

```python
# Understanding my Sentinel image to help problem solve errors

print(ro_s2.name)
print(ro_s2.tags())
print(ro_s2.count)
ro_s2.crs
ro_s2.profile
```

```python
# Clip Sentinel-2 image using mask from rasterio
from rasterio.mask import mask
with rasterio.open(hbg_s2_file) as s2_img:
    s2_clipped, transform = mask(s2_img, hbg_aoi_geometry, crop=True)

# add metadata to new clipped image by copying orginal metadata, then write new GeoTiff
s2_meta = ro_s2.meta.copy()
s2_meta.update(
    {
    
        "transform": transform,
        "height":s2_clipped.shape[1],
        "width":s2_clipped.shape[2]
    }
)
print(s2_meta)    # double check 
with rasterio.open('rwehner_assign5_sentinel2_clipped.tif', 'w', **s2_meta) as my_writer_object:
    my_writer_object.write(s2_clipped)
    
print('Writing complete')    # know when finished
```

```python
# Clip Planet image using mask from rasterio following same steps as previous
with rasterio.open(hbg_planet_file) as p_img:
    p_clipped, transform = mask(p_img, hbg_aoi_geometry, crop=True)

# add metadata to new clipped image by copying orginal metadata, then write new GeoTiff
p_meta = ro_planet.meta.copy()
p_meta.update(
    {
    
        "transform": transform,
        "height":p_clipped.shape[1],
        "width":p_clipped.shape[2]
    }
)
print(p_meta)    # double check
with rasterio.open('rwehner_assign5_planet_clipped.tif', 'w', **p_meta) as my_writer_object:
    my_writer_object.write(p_clipped)
    
print('Writing complete')    # know when finished
```
