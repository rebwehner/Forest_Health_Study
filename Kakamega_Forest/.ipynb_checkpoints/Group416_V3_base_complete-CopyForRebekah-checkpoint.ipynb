```python
# modules needed for processing 

import sentinelsat
from sentinelsat import SentinelAPI, read_geojson, geojson_to_wkt
from datetime import date
import os
import rasterio
from rasterio.mask import mask
from matplotlib import pyplot as plt 
import numpy as np
import geopandas as gp
import zipfile
import pandas as pd
```

```python
api = SentinelAPI(
    'rwehner', 
    '7yN@*bWEd518', 
    'https://apihub.copernicus.eu/apihub'
)
#api

# polygon ID, with defining geometry looking at Kebler Pass, CO
aspen_geometry = {
  "type": "Polygon",
  "coordinates": [
    [ 
      [-107.24182299123405, 38.917511697304235],
      [-107.24182299123405, 38.89846640306999],
      [-107.21691296305106, 38.89846640306999],
      [-107.21691296305106, 38.917511697304235],
      [-107.24182299123405, 38.917511697304235]
    ]
  ]
}

# conversion to readable file for sentinel
aspen_footprint = geojson_to_wkt(aspen_geometry)
```

```python
# command to search sentinel 2 API
# use a large date range where low cloud cover can be selected in desired time series needed fall color spectrum
# determined that peak season for Aspen AoI in 2021 was at the beginning of October and colors will typically last 8 days
# typical fall peak is 9/22-10/3 but for 2021 a warm fall pushed the peak to the end of range

aspen_f_image_metadata = api.query(
    aspen_footprint,
    date=('20210901', '20211020'), # date range of Sept to mid Oct 2021
    platformname='Sentinel-2',  
    processinglevel = 'Level-2A' # image is processed for surface reflection (not Level-1C top of atmosphere)
)

aspen_f_image_metadata_df = api.to_dataframe(aspen_f_image_metadata)
aspen_f_image_metadata_df.to_csv('aspen_full_metadata.csv')

metadata_full_df = api.to_dataframe(aspen_f_image_metadata)

# sorting the data frame to have the clouds with the least cloud cover at the top of the list

aspen_f_image_metadata_df_sorted = aspen_f_image_metadata_df.sort_values(
    ['cloudcoverpercentage'], ascending=[True])

aspen_f_image_metadata_df_sorted.to_csv('aspen_full_metadata_sorted.csv')    # printing to csv to inspect meta values
```

```python
# Define AoI as Kebler Pass, CO, for clipping images
# converting polygon from geojson.io to UTM of image with epsg.io
kp_aoi_geometry = [{
	"type": "Polygon", 
	"coordinates": [ 
	  [
		[305054.11263451225, 4308447.316292203],
		[305002.0328069356, 4306333.528506731],
		[307162.84084437124, 4306280.569739991],
		[307214.3453553761, 4308394.347656664],
		[305054.11263451225, 4308447.316292203]
	  ],
	]
  }]
kp_aoi_geometry    #test
```

```python
# for loop to automate image processing
# first step is to make sure all images that need cropping are in the same folder as the notebook file
#counter for "for-loop"
counter = 1

for f in os.listdir("."):
    if f.endswith(".jp2"):
        
        # "f" is the jp2 file in this for loop, it is processed in a typical manner for the clip
        # assignment of f to variable, as well as opening the jp2 file as a rasterio object 
        image_file = f
        my_raster_image = rasterio.open(f)
        
        # opening the image as rasterio object again for facilitaition of clipping
        with rasterio.open(image_file) as img:
            clipped, transform = mask(img, kp_aoi_geometry, crop=True)
        
        # copying meta from orginal image
        meta = my_raster_image.meta.copy()    
        
        # updating metadata with new data from the clipped image
        meta.update(    
        {
    
        "transform": transform,
        "height":clipped.shape[1],
        "width":clipped.shape[2]
        }
        )
        
        # writing the clipped image to a tif file and using the counter to ensure that the file is not overwritten 
        # change letter 'aspen(X)' depending on position in time series     
        with rasterio.open('aspenZ'+str(counter)+'.tif', 'w', **meta) as my_writer_object:
            my_writer_object.write(clipped)
            print('Writing complete')
            
        counter = counter + 1   
        my_raster_image.close()

        # the loop should print writing complete for however many bands are processed, ie:4 band images = "Writing complete" * 4
        # rerun code with next batch of .jp2 images for a total of 5 times (labeled V,W,X,Y,Z) for each of the 5 dates
```

```python
# assign empty dictionary

banddict = {}

# begin new counter

bandcounter = 1

# for loop to automate process of extracting band arrays from each clipped image
    
for i in os.listdir("."):
    if i.endswith(".tif"):
        
        # pulls tif file as "imgfile"
        
        imgfile = i
        
        # opening the tif as a rasterio object
        
        imgband = rasterio.open(imgfile)
        
        # reading the "first" band since it is a single band image
        
        band = imgband.read([1])
        
       # banddict['variable_'+str(bandcounter)] = "Band_Image_" + str(bandcounter)   <----- please ignore whatever this bs is

        # this appends the dictionary with the iteration of the for-loop, and the the full array from the tif images

        banddict[str(bandcounter)] = band
        
        # increase counter with each for-loop
        
        bandcounter = bandcounter + 1 
        
        

blue = banddict['1']

green = banddict['2']

red = banddict['3']

NIR = banddict['4']


# Example of band algebra with newly exported bands
# Production of NDVI array

NDVI = (NIR-red)/(NIR+red)

NDVI = np.nan_to_num(NDVI)

NDVI
```
